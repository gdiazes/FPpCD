```python
# Otra forma de leer archivos
print("------------------------------------------")
with open("/ejemplo1.txt","r") as archivo:
    contenido = archivo.read()
    print(contenido)

print(archivo.closed)
print(contenido)
```

```python
# Lectura por fragmentos
print("------------------------------------------")
with open("/ejemplo1.txt","r") as archivo:
    print(archivo.read(4))

print("------------------------------------------")
with open("/ejemplo1.txt","r") as archivo:
    print(archivo.read(4))
    print(archivo.read(4))
    print(archivo.read(4))
    print(archivo.read(4))
```

```python
# lectura por línea
print("------------------------------------------")
with open("/ejemplo1.txt","r") as archivo:
    print("Primera línea:",archivo.readline())

# lectura de todas las líneas
print("------------------------------------------")
with open("/ejemplo1.txt","r") as archivo:
    for linea in archivo:
        print("Línea:",linea)

# lectura de todas las líneas en una lista
print("------------------------------------------")
with open("/ejemplo1.txt","r") as archivo:
    contenido = archivo.readlines()
    print("Contenido:",contenido)

for linea in contenido:
    print(linea)
```

## Escritura de archivos

```python
# Escritura en un archivo
print("------------------------------------------")
with open("/ejemplo2.txt","w") as archivo:
    archivo.write("Es una nueva línea")
```

```python
# Escritura de más líneas en un archivo
print("------------------------------------------")
with open("/ejemplo2.txt","w") as archivo:
    archivo.write("Es una nueva línea 1")
    archivo.write("Es otra nueva línea 2")
```

```python
# Escritura de más líneas en un archivo
print("------------------------------------------")
with open("/ejemplo2.txt","w") as archivo:
    archivo.write("Es una nueva línea \n")
    archivo.write("Es otra nueva línea \n")
```

```python
# Agregar una línea en un archivo
print("------------------------------------------")
with open("/ejemplo2.txt","a") as archivo:
    archivo.write("Es otra nueva línea agregada \n")
```

```python
# Copiar un archivo
print("------------------------------------------")
with open("/ejemplo2.txt","r") as archivo_origen:
    with open("/ejemplo3.txt","w") as archivo_destino:
        for linea in archivo_origen:
            archivo_destino.write(linea)
```
